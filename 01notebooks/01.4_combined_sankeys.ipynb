```python
# import necessary libraries
import pandas as pd
import plotly.graph_objects as go
```

# Combined Sankey

```python
# Load the CSV
df_combined = pd.read_csv("/Users/charlottewestenberg/Thesis/00data/from_excel/total sankey - policy cov.csv")

# create unique node list
all_nodes = pd.concat([df_combined["Source"], df_combined["Target"]]).unique()
node_dict = {node: i for i, node in enumerate(all_nodes)}

# map to numeric IDs
df_combined["source_id"] = df_combined["Source"].map(node_dict)
df_combined["target_id"] = df_combined["Target"].map(node_dict)


carrier_colors = {
    "Limestone (Manufacturing)": "rgba(120, 140, 190, 0.85)",  # clearer blue (industrial)
    "Limestone (Aggregates)":    "rgba(200, 185, 140, 0.80)",  # sandy beige (construction feel)
    "Mixed Mineral":             "rgba(150, 140, 130, 0.75)",  # neutral warm grey (mix)

    "Oil":                "rgba(20, 20, 20, 0.9)",        # darker → strong black in greyscale
    "Gas":                "rgba(255, 165, 0, 0.85)",      # keep orange, slightly अधिक opaque
    "Coal":               "rgba(70, 50, 35, 0.9)",        # darker brown → separates from refinery
    "Refinery Products":  "rgba(170, 120, 70, 0.8)",      # slightly lighter than before
    "Fossil materials":   "rgba(210, 180, 120, 0.6)",     # lighter + more transparent
    "Carbon Emissions":   "rgba(200, 30, 30, 0.9)",        # darker red → strong contrast in grey

    "Crops": "rgba(140, 210, 90, 0.85)",     # lighter → clearly light grey
    "Wood": "rgba(30, 80, 30, 0.95)",        # darker → strong dark anchor
    "Animal Products": "rgba(200, 160, 80, 0.85)",  # slightly darker ochre
    "Products from biomass": "rgba(40, 140, 160, 0.85)",  # slightly darker teal
    "Mixed Biogenic": "rgba(130, 130, 130, 0.65)",  # pushed toward neutral grey
    "Carbon Emissions (Biogenic)": "rgba(210, 90, 30, 0.9)",  # darker orange → stronger contrast

    "Mixed Material":            "rgba(170, 170, 170, 0.7)",   # neutral grey (for mixed/uncategorized)
}

df_combined["color"] = df_combined["Carrier"].map(carrier_colors).fillna("rgba(200,200,200,0.4)")

#calculate outflows per source and target nodes
source_totals = df_combined.groupby("Source")["Value_tC"].sum()
node_inflow = df_combined.groupby("Target")["Value_tC"].sum()

labels_with_values = []
for node in all_nodes:
    inflow = node_inflow.get(node, 0)
    outflow = source_totals.get(node, 0)
    # choose what to display
    value = inflow if inflow > 0 else outflow
    if value > 0:
        labels_with_values.append(f"{node}<br>{value/1e6:.1f} MtC")
    else:
        labels_with_values.append(node)

fig = go.Figure(data=[go.Sankey(
    
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels_with_values,
        color="lightgrey"
    ),

    link=dict(
        source=df_combined["source_id"],
        target=df_combined["target_id"],
        value=df_combined["Value_tC"],
        color=df_combined["color"]
    )
)])

for carrier, color in carrier_colors.items():
    fig.add_trace(go.Scatter(
        x=[None],
        y=[None],
        mode='markers',
        marker=dict(size=10, color=color),
        legendgroup=carrier,
        showlegend=True,
        name=carrier
    ))

fig.update_layout(
    title_text="EU Carbon Flows (2023)",
    font_size=16,
    width=1600,
    height=1000,
    showlegend=True,
    legend=dict(
        title="Energy carrier",
        orientation="h",
        y=-0.1
    ),
    xaxis=dict(visible=False),
    yaxis=dict(visible=False),
    plot_bgcolor="white",
    paper_bgcolor="white"
)

fig.show()
```

```python
# Load the CSV
df_combined = pd.read_csv("/Users/charlottewestenberg/Thesis/00data/from_excel/total sankey - policy cov.csv")

# create unique node list
all_nodes = pd.concat([df_combined["Source"], df_combined["Target"]]).unique()
node_dict = {node: i for i, node in enumerate(all_nodes)}

# map to numeric IDs
df_combined["source_id"] = df_combined["Source"].map(node_dict)
df_combined["target_id"] = df_combined["Target"].map(node_dict)

policy_colors = {
    "yes":   "rgba(220, 50, 50, 0.75)",  # bright red for emissions
    "no":            "rgba(170, 170, 170, 0.7)",   # neutral grey (for mixed/uncategorized)
}

df_combined["color"] = df_combined["policy_cov"].map(policy_colors).fillna("rgba(200,200,200,0.4)")

#calculate outflows per source and target nodes
source_totals = df_combined.groupby("Source")["Value_tC"].sum()
node_inflow = df_combined.groupby("Target")["Value_tC"].sum()

#group policies per node instead of flow
from collections import defaultdict
node_policies = defaultdict(set)

for _, row in df_combined.iterrows():
    if pd.notna(row["policy_name"]):
        policies = str(row["policy_name"]).split(",")
        policies = [p.strip() for p in policies if p.strip()]
        for policy in policies:
            node_policies[row["Target"]].add(policy)

excluded_node = "Carbon Emissions"

labels_with_values = []
for node in all_nodes:
    policies = node_policies.get(node, set())
    
    if node == excluded_node:
        labels_with_values.append(f"<b>{node}</b>")
        continue

    if policies:
        policy_text = "<br>".join(sorted(policies))
        labels_with_values.append(f"<b>{node}</b><br>{policy_text}")
    else:
        labels_with_values.append(f"<b>{node}</b>")

fig = go.Figure(data=[go.Sankey(
    
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels_with_values,
        color="lightgrey"
    ),

    link=dict(
        source=df_combined["source_id"],
        target=df_combined["target_id"],
        value=df_combined["Value_tC"],
        color=df_combined["color"]
    )
)])

for policy, color in policy_colors.items():
    fig.add_trace(go.Scatter(
        x=[None],
        y=[None],
        mode='markers',
        marker=dict(size=10, color=color),
        legendgroup=policy,
        showlegend=True,
        name=policy
    ))

fig.update_layout(
    title_text="EU Carbon Flows (2023)",
    font_size=16,
    width=1500,
    height=1400,
    showlegend=True,
    legend=dict(
        title="Policy coverage",
        orientation="h",
        y=-0.1
    ),
    xaxis=dict(visible=False),
    yaxis=dict(visible=False),
    plot_bgcolor="white",
    paper_bgcolor="white"
)

fig.show()
```

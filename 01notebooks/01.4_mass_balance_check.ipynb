```python
def check_balance(df):
    inflow = df[df['Type'] == 'Inflow']['Carbon_Mass'].sum()
    outflow = df[df['Type'] == 'Outflow']['Carbon_Mass'].sum()
    stock = df[df['Type'] == 'Stock']['Carbon_Mass'].sum()
    
    balance = inflow - (outflow + stock)
    
    if abs(balance) < 0.01: # Allowing for rounding errors
        print("Balance Closed! Mass is conserved.")
    else:
        print(f"Error: Missing {balance} tonnes of carbon!")
```

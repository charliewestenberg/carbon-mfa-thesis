```python
import pandas as pd

# 1. 'Library' of CCFs
ccf_library = {
    "Crude oil": 0.85,
    "Lignite_EU": 0.276,
    "Petroleum coke": 0.936,
    "Natural gas": 0.734
}

# 2. The Automation Function
def apply_ccf(row):
    product = row['Carrier']
    mass = row['Amount_kg']
    factor = ccf_library.get(product, 0) # Gets the CCF or returns 0 if missing
    return mass * factor

# one line of code calculates 'Carbon_Mass' column

# df['Carbon_Mass'] = df.apply(apply_ccf, axis=1)
```

```python
import pandas as pd

# 1. Load your CSV (assuming you saved the CCF tab as 'ccf_data.csv')
# If you stick to Excel, use: pd.read_excel('Beschikbare Datasets.xlsx', sheet_name='CCF')
df_ccf = pd.read_csv('CCF.csv')

# 2. Automation: Create a dictionary of CCFs automatically
# We filter out the 'Header' rows (like "Fossil") and rows you haven't filled yet
ccf_library = df_ccf.dropna(subset=['Carbon content factor [kg C/kg]'])
ccf_library = dict(zip(ccf_library['Carrier'], ccf_library['Carbon content factor [kg C/kg]']))

# Print it to see your "Library"
print("Your Automated CCF Library:")
for carrier, value in ccf_library.items():
    print(f" - {carrier}: {value} kg C/kg")
```

```python
# Assuming you saved your 'Sectors' tab as 'sectors.csv
df_sectors = pd.read_csv('sectors.csv')

# Let's say you have a column 'Amount_tonnes' that you filled with Eurostat data
def calculate_carbon_mass(row):
    # This automatically looks up the factor from your dictionary!
    factor = ccf_library.get(row['Flow'], 0) 
    return row['Amount_tonnes'] * factor

# Create the new column
df_sectors['Carbon_Mass'] = df_sectors.apply(calculate_carbon_mass, axis=1)

# AUTOMATED SUMS
inflows = df_sectors[df_sectors['Inflow / outflow / stock'] == 'Inflow']['Carbon_Mass'].sum()
outflows = df_sectors[df_sectors['Inflow / outflow / stock'] == 'Outflow']['Carbon_Mass'].sum()
stocks = df_sectors[df_sectors['Inflow / outflow / stock'] == 'Stock']['Carbon_Mass'].sum()

print(f"Total Inflow: {round(inflows, 2)} Mt C")
print(f"Total Outflow + Stock: {round(outflows + stocks, 2)} Mt C")
print(f"System Balance Error: {round(inflows - (outflows + stocks), 2)} Mt C")
```
